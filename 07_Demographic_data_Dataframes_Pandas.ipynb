**Introduction to Seaborn**

```python
#matplot lib is for basic visualization
#seaborn is for high end visualization
#pandas is for data analytics of the data set
#numpy is to perforn numerical or statistical function in datas
#scikitlearn is for alla the most machine learning algorithm

from pandas.errors import InvalidIndexError
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize']=8,8
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
#12.27
visual1 = sns.distplot(stats['InternetUsers'],bins=40)
```

```python
from matplotlib import figure
#boxplot
visualbox2=sns.boxplot(data=stats, x='IncomeGroup', y='BirthRate')
plt.rcParams['figure.figsize'] = 20,10
```

```python
#lmplot means linear modal plot
visualbox3=sns.lmplot(data=stats, x='InternetUsers', y='BirthRate')
```

```python
#removed regression line
visualbox3=sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', fit_reg=False)
```

```python
#hue refers to color code to differentiate the colors of each income group
visualbox3=sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', fit_reg=False, hue='IncomeGroup')
```

```python
visualbox3=sns.scatterplot(data=stats, x='InternetUsers', y='BirthRate', hue='IncomeGroup')
```
